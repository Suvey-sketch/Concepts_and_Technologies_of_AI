Task1: Sum of Nested Lists:

```python
def sum_nested_list(nested_list):
  total = 0
  for element in nested_list:
    if isinstance(element, list):
      total += sum_nested_list(element)
    else:
      total += element
  return total
```

```python
nested_list = (1,2,[2,3],[5,6,4],7,8,[9,2,3,1])
result = sum_nested_list(nested_list)
print("Sum of nested list:", result)
```

```python
def generate_permutation(s):
  permutation = []

  if len(s) <= 1:
    return[s]

  for i in range(len(s)):
    current_char = s[i]
    remaining_char = s[:i] + s[i+1:]

    for perm in generate_permutation(remaining_char):
      new_perm = current_char + perm
      permutation.append(new_perm)
  return permutation
```

```python
print(generate_permutation("abc"))
print(generate_permutation("aab"))
```

```python
def calculate_directory_size(directory_structure):
  total = 0

  for key, value in directory_structure.items():
    if isinstance(value, dict):
      total += calculate_directory_size(value)
    if isinstance(value, int):
      total += value

  return total
```

```python
directory_structure ={
    'file1.txt': 200,
    'file2.txt': 300,
    "subdir1": {
        'file3.txt': 400,
        'file4.txt': 100,
    },
    "subdir2": {
        "subsubdir1": {
            'file5.txt': 250,
        },
        "file6.txt":150
    }
}
result = calculate_directory_size(directory_structure)
print("total size: ", result)
```

Task 1 - Coin Change Problem:

```python
def min_coins(coins,amount):
  if amount == 0:
    return 0
  dp = [float('inf')] * (amount + 1)
  dp[0] = 0
  for coin in coins:
    for i in range(coin, amount+1):
      dp[i] = min(dp[i], dp[i-coin]+1)

  return dp[amount] if dp[amount] != float('inf') else -1
```

```python
print(min_coins([1,2,5], 11))
print(min_coins([5,5,1], 5))
```

Task2 : Longesr Common Subsequence(LCS)

```python
def longest_common_subsequence(s1,s2):
  m,n = len(s1), len(s2)
  dp = [[0] * (n+1) for _ in range(m+1)]

  for i in range(1, m+1):
    for j in range(1, n+1):
      if s1[i-1] == s2[j-1]:
        dp[i][j] = dp[i-1][j-1] + 1
      else:
        dp[i][j] = max(dp[i-1][j], dp[i][j-1])
  return dp[m][n]
```

```python
print(longest_common_subsequence("abcde", "ace"))
```

Task 3 - 0/1 Knapsack Problem:

```python
def knapsack(weights, values, capacity):
  n = len(weights)
  dp = [[0] * (capacity + 1) for _ in range(n+1)]

  for i in range(1, n+1):
    for j in range(1, capacity+1):
      if weights[i-1] <= j:
        dp[i][j] = max(values[i-1] + dp[i-1][j - weights[i-1]], dp[i-1][j])
      else:
        dp[i][j] = dp[i-1][j]
  return dp[n][capacity]
```

```python
weights = [1,3,4,5]
values =[1,4,5,7]
capacity = 7
print(knapsack(weights, values, capacity))
```
