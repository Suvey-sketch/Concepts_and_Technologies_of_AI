Problem 4: Matrix Operations:


```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
```

1. Prove A.A-1 = I

```python
A_inverse = np.linalg.inv(A)
identity = np.dot(A,A_inverse)
print("A.A-inverse = I : \n", identity)
```

2. Prove AB/=BA

```python
AB = np.dot(A,B)
BA = np.dot(B,A)
print("AB: \n", AB)
print("BA: \n", BA)
print("Is AB = BA : ",np.array_equal(AB,BA))
```

Prove (AB)T = BTAT

```python
AB_transpose = AB.T
B_transpose = B.T
A_transpose = A.T
AT_BT = np.dot(A.T,B.T)
print("AB transpose: \n", AB_transpose)
print("A transpose B transpose: \n", AT_BT)
print("Is ABT = AT.BT : ",np.array_equal(AB_transpose,AT_BT))
```

• Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1

x − y + 2z = −3

3x + y − z = 9

```python
A = np.array([[2,-3,1],[1,-1,2],[3,1,-1]])
B = np.array([-1,-3,9])
```

```python
A_inv = np.linalg.inv(A)
X = np.dot(A_inv,B)
print("X[x,y,z]= ",X)
```
