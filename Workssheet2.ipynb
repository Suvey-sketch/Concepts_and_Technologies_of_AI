```python
medical_df = pd.read_csv("/content/drive/MyDrive/Concepts and Technologies of AI/medical_students_dataset.csv")
```

```python
medical_df.head()
```

```python
medical_df.isnull().sum()
```

```python
for col in medical_df.columns:
  null_count = medical_df[col].isnull().sum()
  total = len(medical_df[col])

  if null_count == total:
    print("Column: ", col, "has all null values")

    if medical_df[col].dtype != 'object':
      medical_df[col] = medical_df[col].fillna(0)
    else:
      medical_df[col] = medical_df[col].fillna("NA")

  elif null_count > 0:
    if medical_df[col].dtype != 'object':
      medical_df[col] = medical_df[col].fillna(medical_df[col].median())
      print("filled column: ", col, "with median.")
    else:
      medical_df[col] = medical_df[col].fillna(medical_df[col].mode())
      print("filled column: ", col, "with mode.")
```

```python
medical_df.duplicated().sum()
```

```python
medical_df.shape
```

```python
medical_df = medical_df.drop_duplicates()
```

**3.2** Exercises - Data Cleaning and Transformations with Titanic Dataset:

```python
df_titanic = pd.read_csv('/content/drive/MyDrive/Concepts and Technologies of AI/Titanic-Dataset.csv')
```

```python
subset_df = df_titanic[['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']]
```

```python
sub_first_class = subset_df[subset_df['Pclass'] == 1]
sub_first_class.head()
```

```python
mean_value = sub_first_class['Fare'].mean()
print("Mean Value of Fare = ", mean_value)
median_value = sub_first_class['Fare'].median()
print("Median value of Fare = ",median_value)
max_value = sub_first_class['Fare'].max()
print("Maximum Value of Fare = ",max_value)
min_value = sub_first_class['Fare'].min()
print("Minimum Value of Fare", min_value)
```

```python
print("Total null value = ",sub_first_class['Age'].isnull().sum())
sub_first_class.shape
```

```python
sub_first_class = sub_first_class.dropna(subset = ['Age'])
print("Null value set dropped.")
sub_first_class.shape
```

```python
embarked_dummies = pd.get_dummies(df_titanic['Embarked'], prefix='Embarked')
```

```python
embarked_dummies_df = df_titanic.copy()
embarked_dummies_df = pd.concat([embarked_dummies_df, embarked_dummies], axis=1)
```

```python
embarked_dummies_df = embarked_dummies_df.drop(['Embarked'], axis=1)
embarked_dummies_df = pd.concat([embarked_dummies_df, embarked_dummies], axis=1)
print(embarked_dummies_df[['Name', 'Embarked_C', 'Embarked_Q', 'Embarked_S']].head(10))
print("DF Shape = ", embarked_dummies_df.shape)
```

```python
gender_survive = df_titanic.groupby('Sex')['Survived'].mean()
print(gender_survive)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x= gender_survive.index, y = gender_survive.values)
plt.xlabel('Gender')
plt.ylabel('Survived Rate')
plt.title('Survived Rate by Gender')
plt.show()
```
